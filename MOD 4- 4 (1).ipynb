```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data=tele,x=tele["MonthlyRevenue"])
plt.title("Boxplot of a sample telecom dataset ") #WE ARE USING BOXPLOT ANALYSIS TO DETECT OUR OUTLIERS
```

```python
outlier1 = tele[tele['MonthlyRevenue']> 216].copy()
print((outlier1['CustomerID']))
#IN THIS WE ARE FILTERING OUT THE MONTHLY REVENUE OUTLIER WHOSE VALUES ARE ABOVE 216 AS IT IS THE 
  #STANDARD REGULATION BY THE TELECOM GOV RULES
```

```python
def boxplot(column):
    sns.boxplot(data=tele,x=tele[f"{column}"])
    plt.title(f"Boxplot of Telecom Industry with varied values {column}")
    plt.show()
#WE ARE NOW CREATING THE BOXPLOTS OF THE RELEVANT OUTLIERS WRT TO OUR MONTHLY REVENUES
#ABOVE 216 AND GET THE BEST OUTLIERS FOR OUR ANALYSIS WITH OUR BOXPLOTS
```

```python
boxplot('MonthlyRevenue')
boxplot('MonthlyMinutes')
boxplot('TotalRecurringCharge')
boxplot('DirectorAssistedCalls')
boxplot('OverageMinutes')
boxplot('RoamingCalls')
boxplot('PercChangeMinutes')
boxplot('PercChangeRevenues')
boxplot('ReferralsMadeBySubscriber')
boxplot('IncomeGroup')
```
